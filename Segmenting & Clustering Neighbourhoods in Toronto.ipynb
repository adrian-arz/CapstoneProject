## 1. Creating the database

```python
import pandas as pd
import numpy as np
```

We use pandas to import the dataframe, [0] corresponds to the index of the table in the html site.

```python
neigh_tor=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
neigh_tor.head()
```

```python
neigh_tor=neigh_tor[neigh_tor.Borough!='Not assigned']
print(neigh_tor.shape)
```

## 2. Completing the dataframe with location coordinates

```python
!conda install -c conda-forge geopy --yes
from geopy.geocoders import Nominatim 
```

```python
def get_latlong(postal_code):
    lat_long_coords=None
    while(lat_long_coords is None):
        address=postal_code
        geolocator = Nominatim(user_agent=address)
        location = geolocator.geocode(address)
        lat_long_coords=[location.latitude,location.longitude]
    return lat_long_coords

print(get_latlong('M3A'))
    
```
